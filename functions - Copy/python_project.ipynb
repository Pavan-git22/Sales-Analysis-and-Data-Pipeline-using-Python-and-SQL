```python
#read data and treat null values if any
import pandas as pd

df = pd.read_csv("orders.csv", na_values=['Not Available','unknown'])  # change name if needed
df.head(20)
df['Ship Mode'].unique()
```

```python
#rename column name with small letters
# df.rename(columns = {'Order ID':'order_id',
#                      'Customer ID':'customer_id',
#                      'Order Date':'order_date',})
# df.head(20)

# df.columns = df.columns.str.lower()
# df.head(20)

df.columns = df.columns.str.replace(' ', '_')
df.head(20)
```

```python
# derive new columns discount ,  sale price, profit
df['discount']=df['list_price']*df['discount_percent']/100
df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price'] 
df.head(20)
```

```python
# convert order date to datetime format
df['order_date'] = pd.to_datetime(df['order_date'],format='%Y-%m-%d')
df.head(20)
```

```python
# drop cost price list price and discount percent
df.drop(['cost_price','list_price','discount_percent'],axis=1,inplace=True)
df.head(20)
```
